```python
# Se importa el paquete requests para hacer solicitudes HTTP
import requests
# Se importan los paquetes necesarios para visualizar la imagen con matplotlib y manipularla con PIL
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO

# Se define la variable URL con la direcci√≥n web de la API que devuelve una imagen aleatoria de la raza de perro "affenpinscher"
URL = "https://dog.ceo/api/breed/affenpinscher/images/random"

# Se hace una solicitud GET a la API y se almacena la respuesta en la variable "respuesta"
respuesta = requests.get(URL)

# Se imprime el c√≥digo de respuesta HTTP para verificar si la solicitud fue exitosa (el c√≥digo 200 indica √©xito)
print(f"El c√≥digo es {respuesta.status_code} \n")

# Se verifica si el c√≥digo de respuesta es 200 (√©xito)
if respuesta.status_code == 200:
    # Imprime el contenido completo del JSON recibido de la API
    print('Data:', respuesta.json())
    
    # Guarda el JSON de la respuesta en la variable "Data"
    Data = respuesta.json()
    
    # Extrae la URL de la imagen del campo "message" en el JSON
    url = Data['message']
    
    # Realiza otra solicitud GET para obtener la imagen desde la URL que se extrajo
    response = requests.get(url)

    # Abre la imagen usando Pillow y la carga en memoria con BytesIO
    img = Image.open(BytesIO(response.content))
    
    # Muestra la imagen utilizando matplotlib, y oculta los ejes para que solo se vea la imagen
    plt.imshow(img)
    plt.axis('off')  # Ocultar los ejes
    plt.show()

# Si el c√≥digo de respuesta no es 200, imprime un mensaje de error y el contenido de la respuesta
else:
    print('Error en la solicitud, detalles:', response.text)
```
