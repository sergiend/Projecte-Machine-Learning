```python
# Importa la biblioteca 'requests' para realizar solicitudes HTTP.
import requests
# Importa 'BeautifulSoup' de la biblioteca 'bs4' para analizar el contenido HTML.
from bs4 import BeautifulSoup

# Define la URL del sitio web desde el que se extraerán los datos de trabajos falsos.
URL = "https://realpython.github.io/fake-jobs/"
# Realiza una solicitud GET a la URL especificada y almacena la respuesta en 'page'.
page = requests.get(URL)

# Analiza el contenido de la página con BeautifulSoup usando el parser de HTML.
soup = BeautifulSoup(page.content, "html.parser")
# Encuentra el contenedor que tiene el ID 'ResultsContainer', que contiene los resultados de los trabajos.
results = soup.find(id="ResultsContainer")
# Busca todos los elementos 'div' con la clase 'card-content' dentro del contenedor de resultados.
job_elements = results.find_all("div", class_="card-content")

# Itera sobre cada elemento de trabajo encontrado.
for job_element in job_elements:
    # Encuentra el elemento 'h2' con la clase 'title' que contiene el título del trabajo.
    title_element = job_element.find("h2", class_="title")
    # Encuentra el elemento 'h3' con la clase 'company' que contiene el nombre de la empresa.
    company_element = job_element.find("h3", class_="company")
    # Encuentra el elemento 'p' con la clase 'location' que contiene la ubicación del trabajo.
    location_element = job_element.find("p", class_="location")
    # Imprime el texto del título del trabajo, eliminando espacios en blanco al principio y al final.
    print(title_element.text.strip())
    # Imprime el texto del nombre de la empresa, eliminando espacios en blanco al principio y al final.
    print(company_element.text.strip())
    # Imprime el texto de la ubicación del trabajo, eliminando espacios en blanco al principio y al final.
    print(location_element.text.strip())
    # Imprime una línea en blanco para separar los resultados de diferentes trabajos.
    print()
```
